```python
import cv2
import pytesseract
from matplotlib import pyplot as plt
import numpy as np
import re

def show(data):
	plt.xticks([])
	plt.yticks([])
	plt.imshow(data)
	plt.show()
```

```python
img = cv2.imread('santa_jerry.png')
# img = cv2.bitwise_not(img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(img)
```

```python
configuration = ("-l eng --oem 1 --psm 3")
img.shape
```

```python
cropped_image = img[:-1, 220:-1]
show(cropped_image)
```

```python
ret2,threshold_image = cv2.threshold(cropped_image,255,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
show(threshold_image)
data = pytesseract.image_to_string(threshold_image, configuration)
pred = re.sub(r"\n", " ", " ".join(data))
pred = re.sub(r"\t", " ", pred)
pred
```

```python
# kernel = np.array([ [0, -1, 0], [-1, 5,-1], [0, -1, 0] ])
kernel = np.array([ [-1, -1, -1], [-1, 9, -1], [-1, -1, -1] ])
sharpened = cv2.filter2D(cropped_image, -1, kernel)
show(sharpened)
data = pytesseract.image_to_string(sharpened, config=configuration)
pred = re.sub(r"\n", " ", "".join(data))
pred = re.sub(r"\t", " ", pred)
pred
```
