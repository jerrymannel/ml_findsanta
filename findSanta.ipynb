```python
import cv2
import pytesseract
from matplotlib import pyplot as plt
import numpy as np
import re

def show(data):
	plt.xticks([])
	plt.yticks([])
	plt.imshow(data)
	plt.show()
```

```python
img = cv2.imread('2023.png')
# img = cv2.bitwise_not(img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(img)
```

```python
configuration = ("-l eng --oem 1 --psm 3")
img.shape
```

```python
# kernel = np.array([ [0, -1, 0], [-1, 5,-1], [0, -1, 0] ])
kernel = np.array([ [-1, -1, -1], [-1, 9, -1], [-1, -1, -1] ])
sharpened = cv2.filter2D(img, -1, kernel)
show(sharpened)
data = pytesseract.image_to_string(sharpened, config=configuration)
pred = re.sub(r"\n", " ", "".join(data))
pred = re.sub(r"\t", " ", pred)
pred
```
