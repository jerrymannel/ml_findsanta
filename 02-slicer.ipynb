```python
import os
import cv2
from matplotlib import pyplot as plt
import pathlib as path

def show(data):
	plt.xticks([])
	plt.yticks([])
	plt.imshow(data)
	plt.show()

def printArrayInfo(arr):
	print("---------------------------")
	# print(arr)
	print("Size :: ", arr.size)
	print("Shape :: ", arr.shape)
	print("---------------------------")
```

```python
slidesToSkip = [2, 23, 34]

height, width = 2799, 2812
rowHeight = int(height / 15) - 1
colWidth = int(width / 15)
print(rowHeight, colWidth)

pageDirectory = path.PurePath("pages")
data = path.PurePath("data")
dataTest = path.PurePath("dataTest")
```

# Crop the slide

Crop the slide and get only the character grid

```python
def getPageAndCropGrid(page):
    print("Processing page - " + page)
    page = cv2.imread(page)
    croppedPage = page[:-1, 2200:-1]
    return croppedPage
```

# Slice grid into separate characters

```python
def sliceIntoSingleCharacters(basePath, croppedPage):
		x, y = 0, 0
		rowCounter = 0
		imageCounter = 1000
		while x < 15:
				# print(rowCounter, rowCounter + rowHeight)
				row = croppedPage[rowCounter:rowCounter + rowHeight]
				colCounter = 0
				charCounter = 1
				y = 0
				while y < 15:
						col_text = row[:-1, colCounter:colCounter + colWidth]
						imageName = "{}_{}_{}.jpg".format(basePath, x, y)
						cv2.imwrite(imageName, col_text)
						print("File {} saved.".format(imageName))
						colCounter = colCounter + colWidth
						charCounter += 1
						imageCounter += 1
						y += 1
				rowCounter = rowCounter + rowHeight
				x += 1
```

# Start here

```python
for d in os.listdir(pageDirectory):
	if d.count("jpg") == 0:
		continue
	print("Processing page {}".format(d))
	croppedPage = getPageAndCropGrid(str(pageDirectory.joinpath(d)))
	pageName = d.split(".jpg")[0]
	if d.count("TEST") > 0:
		print("Found a test page")
		sliceIntoSingleCharacters(dataTest.joinpath(pageName), croppedPage)
	else:
		print("Found a training page")
		sliceIntoSingleCharacters(data.joinpath(pageName), croppedPage)
```
