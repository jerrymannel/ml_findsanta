```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

def show(data):
	plt.xticks([])
	plt.yticks([])
	plt.imshow(data)
	plt.show()

def printArrayInfo(arr):
	print("---------------------------")
	# print(arr)
	print("Size :: ", arr.size)
	print("Shape :: ", arr.shape)
	print("---------------------------")
```

# Read an image

```python
imageName = "1_14_4"
# imageName = "test_slide_new_34"
image = cv2.imread("dataClassified/a_lower/5_4_10.jpg")
show(image)
```

# Clean Image

```python
greaterThanGrayPixels = np.where((
	(image[:, :, 0] > 200) &
	(image[:, :, 1] > 200) &
	(image[:, :, 2] > 200)
))
image[greaterThanGrayPixels] = [255, 255, 255]

lessThanGrayPixels = np.where((
	(image[:, :, 0] < 200) &
	(image[:, :, 1] < 200) &
	(image[:, :, 2] < 200)
))
image[lessThanGrayPixels] = [0, 0, 0]

show(image)
```

# Generating pixel shifted images

```python
img = image.copy()
num_rows, num_cols = img.shape[:2]
print(num_rows, num_cols)
translation_matrix = np.float32([ [1,0,40], [0,1,20] ])
img_translation = cv2.warpAffine(img, translation_matrix, (num_cols,num_rows), borderMode=cv2.BORDER_CONSTANT, borderValue=(255,255,255))
show(img)
show(img_translation)
```
